```python
import scipy.stats as stats

control_group = user_spending[user_spending['group'] == 'control']['amount'].dropna()
test_group = user_spending[user_spending['group'] == 'treatment']['amount'].dropna()

control_stat, control_p = stats.shapiro(control_group)

test_stat, test_p = stats.shapiro(test_group)

alpha = 0.05

print(control_p)
print(test_p)

if control_p > alpha:
    print("Контрольная группа: Нормальное распределение")
else:
    print("Контрольная группа: Не нормальное распределение")

if test_p > alpha:
    print("Тестовая группа: Нормальное распределение")
else:
    print("Тестовая группа: Не нормальное распределение")
```
