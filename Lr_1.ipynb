```python
df_new = df[df['income'] > 1000].groupby(['title', 'status']).agg({'income': 'sum', 'id': 'count'}).sort_values(by = ['title', 'status'])
```

```python
df_new
```

```python

df_filtered = df[df['income'] >= 1000]

df_grouped = df_filtered.groupby(["city", "title"], as_index=False)
df_grouped_agg = df_grouped.agg({'income': 'sum'})
df_grouped_sorted = df_grouped_agg.sort_values(by = ['city', 'income'], ascending = False)
df_final = df_grouped_sorted.groupby('city').head(3)
df_final.head(15)
```

```python
df_final.to_csv('final.csv')
```
