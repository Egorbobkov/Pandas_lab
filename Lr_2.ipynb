```python
data.columns = data.columns.str.lower().str.replace(' ', '_')
data.head()
```

```python
succesfull_bookings = data[data['is_canceled'] == 0]
top_country = succesfull_bookings['country'].value_counts()
print(top_country.head())
```

```python
count_night = data.groupby('hotel')['stays_total_nights'].mean()
print(count_night.head())
```

```python
difference = data[data['assigned_room_type'] != data['reserved_room_type']]
print(len(difference))
```

```python
succesfull_booking = data[data['is_canceled'] == 0]
succesfull_2016 = succesfull_booking[succesfull_booking['arrival_date_year'] == 2016]
succesfull_pop_mounth = succesfull_2016['arrival_date_month'].value_counts().idxmax()
print('Самый частый месяц 2016:', succesfull_pop_mounth)
```

```python
succesful_booking = data[data['is_canceled'] == 0]
succesful_2017 = succesful_booking[succesful_booking['arrival_date_year'] == 2017]
succesfull_pop_mounth_2 = succesful_2017['arrival_date_month'].value_counts().idxmax()
print('Самый частый месяц 2017:', succesfull_pop_mounth_2)
```

```python
cancel_booking = data[(data['is_canceled'] == 1) & (data['hotel'] == 'City Hotel')]
cancel_City = cancel_booking.groupby(['arrival_date_year', 'arrival_date_month'])['is_canceled'].count()
best_month_cancel = cancel_City.groupby('arrival_date_year').idxmax()
best_month_cancel = best_month_cancel.apply(lambda x: x[1])
best_month_cancel
```

```python
import numpy as np
adults_mean = data['adults'].mean()
children_mean = data['children'].mean()
babies_mean = data['babies'].mean()
means = np.array([adults_mean, children_mean, babies_mean])
max_mean = np.max(means)


print(adults_mean, children_mean, babies_mean, max_mean, sep=', ')
```

```python
data['total_kids'] = data['children'] + data['babies']
popular_hotel = data.groupby('hotel')['total_kids'].mean()
popular_hotel
```

```python
data['has_kids'] = data['total_kids'] > 0
```

```python

churned_clients = data[data['reservation_status'].isin(['Canceled', 'No-Show'])]
total_clients = len(data)
churned_clients_count = len(churned_clients)
churn_rate = (churned_clients_count / total_clients) * 100

print(f"Churn rate: {churn_rate:.2f}%")

churn_rate_by_hotel = churned_clients.groupby('hotel').size() / data.groupby('hotel').size() * 100

print(churn_rate_by_hotel)
```
